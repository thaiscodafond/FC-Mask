## Préparation
On importe les librairies nécessaires, et on pose les variables qui vont nous être utiles.

```python
import cv2
import numpy as np
from keras.models import load_model

## On télécharge le modèle qu'on a créé
model=load_model("./model2-010.model") 

rect_size = 4

## Nous aide à trouver le visage sur la vidéo
haarcascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml') 

## On se connecte à la caméra de l'ordinateur
video = cv2.VideoCapture(0)
```

## Le code


```python

while True:
    (rval, im) = video.read()
    im=cv2.flip(im,1,1) 
    
    rerect_size = cv2.resize(im, (im.shape[1] // rect_size, im.shape[0] // rect_size))
    faces = haarcascade.detectMultiScale(rerect_size)
    
    for f in faces:
        (x, y, w, h) = [v * rect_size for v in f] 
        
        face_img = im[y:y+h, x:x+w]
        rerect_sized=cv2.resize(face_img,(150,150))
        normalized=rerect_sized/255.0
        reshaped=np.reshape(normalized,(1,150,150,3))
        reshaped = np.vstack([reshaped])
        results=model.predict(reshaped)
        
        confidence = round((results[0][1]*100).item(),2)
        
        label=np.argmax(results,axis=1)[0]
        
        if label == 1 :
            result = "Avec un masque"
            color = (0, 255, 0)
            confiance = str(confidence)
            text = "confiance :"+confiance +"%"
            cv2.putText(im, text,(x+2,y+w-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255),2)
        else:
            result = "Sans masque"
            color = (0, 0, 255)

        cv2.rectangle(im,(x,y),(x+w,y+h),color,2)
        cv2.rectangle(im,(x,y-40),(x+w,y),color,-1)
        cv2.putText(im, result, (x+2, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
    cv2.imshow('LIVE',   im)
    key = cv2.waitKey(10)
    
    if key == ord('q'): 
        break
video.release()
cv2.destroyAllWindows()
```
